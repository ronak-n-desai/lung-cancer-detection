```python
import numpy as np
import matplotlib.pyplot as plt

# Data
categories = ['CNN', 'CNN_Double', 'L2_regularizaton', 'Drop_out', 'BatchNorm', 'ResNet']
values1 = [0.94, 0.95, 0.93, 0.99, 0.99, 0.92]
values2 = [0.91, 0.92, 0.91, 0.94, 0.94, 0.91]

x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars

yticks = np.arange(0.7, 1.05, 0.02)
# Create the plot
fig, ax = plt.subplots(figsize=(15, 8))
rects1 = ax.bar(x - width/2, values1, width, label='Train', color='blue')
rects2 = ax.bar(x + width/2, values2, width, label='Val', color='red')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy %', fontsize = 20)
ax.set_title('Results', fontsize = 20)
ax.set_xticks(x)

ax.set_yticks(yticks)
ax.set_xticklabels(categories, fontsize = 16)
ax.set_ylim(0.80, 1.0)
# Set font size for x and y ticks
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

for bar in rects1:
    yval = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # X position: center of the bar
        yval + 0.02,  # Y position: just above the bar
        round(yval, 2),  # Text value (rounded to 2 decimal places)
        ha='center',  # Horizontal alignment: center
        va='bottom',  # Vertical alignment: bottom (to place text above the bar)
        fontsize=12,  # Font size of the text
        color='blue',  # Text color
        fontweight='bold',  # Font weight
        bbox=dict(facecolor='white', alpha=0.5)  # Background color with transparency
    )
for bar in rects2:
    yval = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # X position: center of the bar
        yval + 0.02,  # Y position: just above the bar
        round(yval, 2),  # Text value (rounded to 2 decimal places)
        ha='center',  # Horizontal alignment: center
        va='bottom',  # Vertical alignment: bottom (to place text above the bar)
        fontsize=12,  # Font size of the text
        color='blue',  # Text color
        fontweight='bold',  # Font weight
        bbox=dict(facecolor='white', alpha=0.5)  # Background color with transparency
    )
ax.legend(loc='lower right',fontsize=16)
# Display the plot
plt.savefig('/kaggle/working/accuracy.png')
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['CNN', 'CNN_Wide', 'L2_reg', 'Drop_out', 'BatchNorm', 'ResNet']
values1 = [0.94, 0.95, 0.93, 0.99, 0.99, 0.92]
values2 = [0.91, 0.92, 0.91, 0.94, 0.94, 0.91]

factor = 100
values1 = [num * factor for num in values1]
values2 = [num * factor for num in values2]
print(values1)
x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars

yticks = np.arange(70, 102, 2)
# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2,  values1, width, label='Train', color='blue')
rects2 = ax.bar(x + width/2, values2, width, label='Val', color='red')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy (%)', fontsize = 20)
ax.set_title('Comparison of Accuracy', fontsize = 20)
ax.set_xticks(x)

ax.set_yticks(yticks)
ax.set_xticklabels(categories, fontsize = 16)
ax.set_ylim(80, 100)
# Set font size for x and y ticks
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xticks(rotation=30)

ax.legend(loc='upper right',fontsize=16)
# Display the plot
plt.savefig('/kaggle/working/accuracy.png', bbox_inches='tight', dpi=300)
```

```python
import pandas as pd

df = pd.read_csv('/kaggle/input/result/BaseCNN_results.csv')
df1 = pd.read_csv('/kaggle/input/result/CNNWide_results.csv')
df2 = pd.read_csv('/kaggle/input/result/L2reg_results.csv')
df3 = pd.read_csv('/kaggle/input/result/Dropout_results.csv')
df4 = pd.read_csv('/kaggle/input/result/Batchnorm_results.csv')
df5 = pd.read_csv('/kaggle/input/result/Rsnt_results.csv')

print(df2.tail())
```
